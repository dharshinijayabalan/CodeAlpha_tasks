```python
def chatbot():
    print("Hello! I'm a simple chatbot.")
    print("You can type: hello, how are you, or bye to end the chat.")

    while True:
        user_input = input("You: ")
        user_input = user_input.lower()

        if user_input == "hello":
            print("Chatbot: Hi")
        elif user_input == "how are you":
            print("Chatbot: I'm fine, thanks!")
        elif user_input == "bye":
            print("Chatbot: Goodbye!")
            break
        else:
            print("Chatbot: I don't understand that")

chatbot()
```
